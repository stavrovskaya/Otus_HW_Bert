```python
from transformers import AutoConfig, AutoModelForCausalLM, AutoTokenizer, TextGenerationPipeline
model_dir = "ai-forever/rugpt3small_based_on_gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_dir)
model = AutoModelForCausalLM.from_pretrained().cuda(model_dir)
```

```python
pipeline = TextGenerationPipeline(model=model, tokenizer=tokenizer, max_new_tokens=200)
```

```python
prompts = {
    "zero_shot": 'Классифицируй это предложение как правильное или неправильное: "Этим летом не никуда ездили."\nРезультат: ',
    "one_shot": 'Классифицируй это предложение как правильное или неправильное: \nПример:\nПредложение: "Этим летом не никуда ездили."\nКлассификация: неправильное\n\nПредложение: "На поверку вся теория оказалась полной чепухой."\nКлассификация: ',
    "two_shot": 'Классифицируй это предложение как правильное или неправильное: \nПример:\nПредложение: "Этим летом не никуда ездили."\nКлассификация: неправильное\n\nПредложение: "На поверку вся теория оказалась полной чепухой."\nКлассификация: правильное\n\nПредложение: "Он всегда сидит в библиотеке."\nКлассификация: ',
    "four_shot": 'Классифицируй это предложение как правильное или неправильное: \nПример:\nПредложение: "Этим летом не никуда ездили."\nКлассификация: неправильное\n\nПредложение: "На поверку вся теория оказалась полной чепухой."\nКлассификация: правильное\n\nПредложение: "Он всегда сидит в библиотеке."\nКлассификация: правильное\n\nПример 3:\nПредложение: "Мне предоставилась возможность все видеть, сам оставаясь незамеченным."\nКлассификация: неправильное\n\nПример 4:\nПредложение: "Наш новый сотрудник оказался коренастым мужчиной лет сорока."\nКлассификация: правильное\n\nПредложение: "Не думаю, что мосты уже сняли."\nКлассификация: '
}
```

```python
answer=pipeline([prompts['zero_shot']])
print(answer)
answer=pipeline([prompts['one_shot']])
print(answer)
answer=pipeline([prompts['two_shot']])
print(answer)
answer=pipeline([prompts['four_shot']])
print(answer)
```
